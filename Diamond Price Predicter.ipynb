```python
def outliers (column):
    q3,q1 = np.percentile(df[column],[75,25])
    iqr = q3 - q1
    lowerOutlier = q1 - 1.5*iqr
    upperOutlier = q3 + 1.5*iqr
    return (lowerOutlier,upperOutlier)
```

```python
dimensions = ['x','y','z','price']
for i in dimensions:
    df_outliers_removed = df[(df[i] < outliers(i)[1] ) & (df[i] > outliers(i)[0])]
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

diamonds_data = pd.read_csv('diamonds.csv')
y = diamonds_data.price
feature_columns = ['carat', 'x', 'y', 'z']
X = diamonds_data[feature_columns]

# Specify Model
diamonds_model = DecisionTreeRegressor()
# Fit Model
diamonds_model.fit(X, y)

print("First in-sample predictions:", diamonds_model.predict(X.head()))
print("Actual target values for those diamonds:", y.head().tolist())
```

```python
# Import the train_test_split function and uncomment
# from _ import _
from sklearn.model_selection import train_test_split

# fill in and uncomment
# train_X, val_X, train_y, val_y = ____
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
```
